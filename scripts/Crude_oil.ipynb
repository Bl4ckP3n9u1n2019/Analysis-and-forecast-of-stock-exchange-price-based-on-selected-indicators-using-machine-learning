```python
#Import data, skip description rows, use columns: date & value
crude_oil = pd.read_csv('wti-crude-oil-prices-10-year-daily-chart.csv',
                     skiprows=14,
                     usecols=['date',' value'])

#Convert column type to date time
crude_oil["date"] = pd.to_datetime(crude_oil["date"])

#Select price between 2010-01-01 and 2020-03-24
filter = (crude_oil['date'] >= '2010-01-01') & (crude_oil['date'] < '2020-03-25')
crude_oil = crude_oil[filter]

#Rename columns
crude_oil = crude_oil.rename(columns={'date': 'Date', ' value': 'Price'})

#Change index column for "Date"
crude_oil.index = crude_oil['Date']

#Delete "Date" column that is not an index
crude_oil = crude_oil.drop(columns=['Date'])

#Display Data Frame
crude_oil.head()
```

## Crude oil price chart

```python
plt.style.use('fivethirtyeight')
crude_oil.plot(title="Crude Oil price chart from 2010-01-01 to 2020-03-25 - source: \"www.macrotrends.net\"",
            figsize=(15,10),grid=True,fontsize=16)
```
