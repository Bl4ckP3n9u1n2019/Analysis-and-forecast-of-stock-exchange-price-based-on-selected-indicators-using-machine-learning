```python
#Import data, skip description rows, use columns: date & value
platinum = pd.read_csv('platinum-prices-historical-chart-data.csv',
                     skiprows=14,
                     usecols=['date',' value'])

#Convert column type to date time
platinum["date"] = pd.to_datetime(platinum["date"])

#Select price between 2010-01-01 and 2020-03-24
filter = (platinum['date'] >= '2010-01-01') & (platinum['date'] < '2020-03-24')
platinum = platinum[filter]

#Rename columns
platinum = platinum.rename(columns={'date': 'Date', ' value': 'Price'})

#Reset the index order
platinum.index = platinum['Date']
platinum = platinum.drop(columns=['Date'])

#Display Data Frame
platinum.head()
```

## Platinum price chart

```python
plt.style.use('fivethirtyeight')
platinum.plot(title="Platinum price chart from 2010-01-01 to 2020-03-24 - source: \"www.macrotrends.net\"",
            figsize=(15,10),grid=True,fontsize=16)
```
