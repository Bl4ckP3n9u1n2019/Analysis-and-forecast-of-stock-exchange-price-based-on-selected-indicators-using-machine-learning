```python
#Display options
pd.set_option('display.min_rows', 100)
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 1000)
```

```python
#Import data, skip description rows, use columns: date & value
palladium = pd.read_csv('palladium-prices-historical-chart-data.csv',
                     skiprows=14,
                     usecols=['date',' value'])

#Convert column type to date time
palladium["date"] = pd.to_datetime(palladium["date"])

#Select price between 2010-01-01 and 2020-03-24
filter = (palladium['date'] >= '2010-01-01') & (palladium['date'] < '2020-03-24')
palladium = palladium[filter]

#Rename columns
palladium = palladium.rename(columns={'date': 'Date', ' value': 'Price'})

#Reset the index order
palladium.index = palladium['Date']
palladium = palladium.drop(columns=['Date'])

#Display Data Frame
palladium.head()
```

## Palladium price chart

```python
plt.style.use('fivethirtyeight')
palladium.plot(title="Palladium price chart from 2010-01-01 to 2020-03-24 - source: \"www.macrotrends.net\"",
            figsize=(15,10),grid=True,fontsize=16)
```
