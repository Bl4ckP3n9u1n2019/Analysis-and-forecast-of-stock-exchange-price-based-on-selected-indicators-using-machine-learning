```python
#Import data, skip description rows, use columns: date & value
natural_gas = pd.read_csv('natural-gas-prices-historical-chart.csv',
                     skiprows=14,
                     usecols=['date',' value'])

#Convert column type to date time
natural_gas["date"] = pd.to_datetime(natural_gas["date"])

#Select price between 2010-01-01 and 2020-03-24
filter = (natural_gas['date'] >= '2010-01-01') & (natural_gas['date'] < '2020-03-25')
natural_gas = natural_gas[filter]

#Rename columns
natural_gas = natural_gas.rename(columns={'date': 'Date', ' value': 'Price'})

#Change index column for "Date"
natural_gas.index = natural_gas['Date']

#Delete "Date" column that is not an index
natural_gas = natural_gas.drop(columns=['Date'])

#Display Data Frame
natural_gas.head()
```

```python
plt.style.use('fivethirtyeight')
natural_gas.plot(title="Natural Gas price chart from 2010-01-01 to 2020-03-25 - source: \"www.macrotrends.net\"",
            figsize=(15,10),grid=True,fontsize=16)
```
